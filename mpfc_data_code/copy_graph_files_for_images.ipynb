#### Loop through all graph files

The file resources are under the folder:

    ../data/mpfc_data/graph_objects_degree_20_prepare

Copy them to the folder

    ../data/mpfc_data/graph_objects_degree_20 

```python
degree_limit = 20
```

```python
target_folder = "../data/mpfc_data/graph_objects_degree_"+str(degree_limit)
source_folder = "../data/mpfc_data/graph_objects_degree_"+str(degree_limit)+"_prepare"
```

```python
target_tg_graph_basic_folder = target_folder+"/tg_graph_basic"    

os.makedirs(target_tg_graph_basic_folder, exist_ok=True)

for region_id in train_images:
    
    source_tg_graph_basic_folder = source_folder+"/"+region_id+"/tg_graph_basic"

    source_tg_graph_basic_path = source_tg_graph_basic_folder+"/"+region_id+".gpt"

    target_tg_graph_basic_path = target_tg_graph_basic_folder+"/"+region_id+".gpt"

    shutil.copyfile(source_tg_graph_basic_path, target_tg_graph_basic_path)
```

```python
target_tg_graph_folder = target_folder+"/tg_graph"    

os.makedirs(target_tg_graph_folder, exist_ok=True)

for region_id in train_images:
    
    source_tg_graph_folder = source_folder+"/"+region_id+"/tg_graph"

    source_tg_graph_path = source_tg_graph_folder+"/"+region_id+".gpt"

    target_tg_graph_path = target_tg_graph_folder+"/"+region_id+".gpt"

    shutil.copyfile(source_tg_graph_path, target_tg_graph_path)
```

```python
target_tg_graph_1st_folder = target_folder+"/tg_graph_1st"    

os.makedirs(target_tg_graph_1st_folder, exist_ok=True)

for region_id in train_images:
    
    source_tg_graph_1st_folder = source_folder+"/"+region_id+"/tg_graph_1st"

    source_tg_graph_1st_path = source_tg_graph_1st_folder+"/"+region_id+".gpt"

    target_tg_graph_1st_path = target_tg_graph_1st_folder+"/"+region_id+".gpt"

    shutil.copyfile(source_tg_graph_1st_path, target_tg_graph_1st_path)
```

```python
target_tg_graph_2nd_folder = target_folder+"/tg_graph_2nd"    

os.makedirs(target_tg_graph_2nd_folder, exist_ok=True)

for region_id in train_images:
    
    source_tg_graph_2nd_folder = source_folder+"/"+region_id+"/tg_graph_2nd"

    source_tg_graph_2nd_path = source_tg_graph_2nd_folder+"/"+region_id+".gpt"

    target_tg_graph_2nd_path = target_tg_graph_2nd_folder+"/"+region_id+".gpt"

    shutil.copyfile(source_tg_graph_2nd_path, target_tg_graph_2nd_path)
```

```python
target_niche_folder = target_folder+"/niche_encoded"
target_niche_ct_group_folder = target_folder+"/niche_ct_group_encoded"

os.makedirs(target_niche_folder, exist_ok=True)
os.makedirs(target_niche_ct_group_folder, exist_ok=True)

for region_id in train_images:
    
    source_niche_folder = source_folder+"/"+region_id+"/niche_encoded"
    source_niche_ct_group_folder = source_folder+"/"+region_id+"/niche_ct_group_encoded"
    
    source_niche_path = source_niche_folder+"/"+region_id+".csv"
    source_niche_ct_group_path = source_niche_ct_group_folder+"/"+region_id+".csv"

    target_niche_path = target_niche_folder+"/"+region_id+".csv"
    target_niche_ct_group_path = target_niche_ct_group_folder+"/"+region_id+".csv"

    shutil.copyfile(source_niche_path, target_niche_path)
    shutil.copyfile(source_niche_ct_group_path, target_niche_ct_group_path)
```

```python
target_composition_folder = target_folder+"/group_composition_2nd_basic"

os.makedirs(target_composition_folder, exist_ok=True)

for region_id in train_images:
    
    source_composition_folder = source_folder+"/"+region_id+"/group_composition_2nd_basic"
    
    source_composition_path = source_composition_folder+"/"+region_id+".csv"
    target_composition_path = target_composition_folder+"/"+region_id+".csv"

    shutil.copyfile(source_composition_path, target_composition_path)
```

```python
target_graph_folder = target_folder+"/graph"    

os.makedirs(target_graph_folder, exist_ok=True)

for region_id in train_images:
    
    source_graph_folder = source_folder+"/"+region_id+"/graph"

    source_graph_path = source_graph_folder+"/"+region_id+".gpkl"

    target_graph_path = target_graph_folder+"/"+region_id+".gpkl"

    shutil.copyfile(source_graph_path, target_graph_path)
```
